```python
# initiate the inventory, these will be the scaling factor for the FU here chosen
fu_ca = bw.Database('PS pilot').get('Use and biowaste') 
lca = bw.LCA({fu_ca: 1},) # contribution analysis for this FU only
lca.lci()
```

```python
# list of all foreground activities
acts = [act['code'] for act in PS_db]
```

```python
# dictionary of scaling factors from supply array
scaling = {}
for act in acts:
    index = lca.activity_dict[('PS pilot', act)] # index is to find the activities in the tech matrix
    scaling[act] = lca.supply_array[index] # this is the amount of activity used, we need it for contribution analysis

scaling
```

```python
# Dictionary of reference flows, because not all are normalized

refflows = {}
for act in acts:
    for exc in list(bw.Database('PS pilot').get(act).exchanges()):
        if exc['type'] == 'production':
            refflows[act] = (exc['amount'])
refflows
```

```python
scaling_norm = dict(zip(acts,[scaling[i]*refflows[i] for i in acts]))
scaling_norm
```

```python
# scaling factors are scaling factors times reference flow
factors = pd.DataFrame([scaling, refflows, scaling_norm], index = ['Scaling','Reference flow', 'Scaling normalized']).T
factors
```

```python
GW = [('IPCC 2013', 'climate change', 'GWP 100a'),
      ('IPCC 2013', 'climate change-biogenic', 'GWP 100a')]
```

```python
def dolcacalc(myact, mydemand, mymethod):
    my_fu = {myact: mydemand} 
    lca = bw.LCA(my_fu, mymethod)
    lca.lci()
    lca.lcia()
    return lca.score

def getLCAresults(acts, mymethod):
    
    all_activities = []
    results = []
    for a in acts:
        act = bw.Database('PS pilot').get(a)
        all_activities.append(act['name'])
        results.append(dolcacalc(act,1,mymethod)) # 1 stays for one unit of each process
        #print(act['name'])
     
    results_dict = dict(zip(all_activities, results))
    
    return results_dict
```

```python
results_GW = []
for m in GW:
    results_all_acts = getLCAresults(acts,m) # total impact per tech
    results_GW.append(results_all_acts)
```

```python
methods_names = []
for m in GW:
    m_name = ' '.join(m)
    methods_names.append(m_name)

my_output = pd.DataFrame(results_GW, index=methods_names)
my_output.T.sort_index().head()
```

```python
scores = pd.DataFrame(my_output.T.sort_index().values * pd.DataFrame(factors['Scaling normalized']).sort_index().values, 
             index = my_output.T.sort_index().index, 
             columns = my_output.index)
scores
```
